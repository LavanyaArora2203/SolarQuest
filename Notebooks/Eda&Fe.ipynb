```python
df1['TimeSunRise'] = pd.to_datetime(
    df1['TimeSunRise'],
    format='%H:%M:%S',
    errors='coerce'
).dt.time

df1['TimeSunSet'] = pd.to_datetime(
    df1['TimeSunSet'],
    format='%H:%M:%S',
    errors='coerce'
).dt.time
```

```python
# Combine date + time
df1['sunrise_dt'] = pd.to_datetime(
    df1['date'].dt.strftime('%Y-%m-%d') + ' ' + df1['TimeSunRise'].astype(str),
    errors='coerce'
)

df1['sunset_dt'] = pd.to_datetime(
    df1['date'].dt.strftime('%Y-%m-%d') + ' ' + df1['TimeSunSet'].astype(str),
    errors='coerce'
)

# Compute daylight duration
df1['daylight_minutes'] = (
    df1['sunset_dt'] - df1['sunrise_dt']
).dt.total_seconds() / 60

```

```python
mask = df1['sunset_dt'] < df1['sunrise_dt']
df1.loc[mask, 'sunset_dt'] += pd.Timedelta(days=1)
```

```python
df_model = df1.drop(
    columns=[
        'UNIXTime', 'time',
        'TimeSunRise', 'TimeSunSet',
        'sunrise_dt', 'sunset_dt',
        'year', 'day'
    ]
)
```

Create STRONG time features (important!)

Solar radiation is cyclical.

```python
df_model['hour'] = df1['date'].dt.hour

import numpy as np
df_model['hour_sin'] = np.sin(2 * np.pi * df_model['hour'] / 24)
df_model['hour_cos'] = np.cos(2 * np.pi * df_model['hour'] / 24)

df_model.drop(columns=['hour'], inplace=True)
```

Remove night-time samples (huge boost)

```python
df_model = df_model[df_model['Radiation'] > 0]
```

```python
df_model = df_model.dropna()
```
